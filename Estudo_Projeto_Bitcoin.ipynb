```python
#importando as Libs
import numpy as np
import pandas as pd

#Libs para plotly
import plotly.express as px
import plotly.graph_objects as Dash
```

```python
#Lendo Base
Base_Dados = pd.read_excel('Dados_Bitcoin.xlsx')

Base_Dados.head(15)
```

```python
# Setar o index
Base_Dados.set_index('Date', inplace = True)
```

```python
# Grafico de linhas

fig = px.line(Base_Dados, y='Close')
fig.show()
```

```python
# Gerar mdias mveis
Media_Movel = Base_Dados['Close'].rolling(5).mean()
Media_Tendencia = Base_Dados['Close'].rolling(30).mean()
```

```python
# Criar Dashboard

# Criando uma Figura
Figura = Dash.Figure()

# Adicionado o primeiro eixo
Figura.add_trace(
    Dash.Scatter(
        x = Base_Dados.index,
        y = Base_Dados.Close,
        mode = 'lines',
        name = 'Fechamento',
        marker_color = '#ff7f0e',
        opacity = 0.5,
    )
)

# Adicionado a mdia
Figura.add_trace(
    Dash.Scatter(
        x = Base_Dados.index,
        y = Media_Movel,
        mode = 'lines',
        name = 'Mdia Mvel',
        marker_color = '#d62728',
        opacity = 0.5,
    )
)

# Adicionado a Tendencia
Figura.add_trace(
    Dash.Scatter(
        x = Base_Dados.index,
        y = Media_Tendencia,
        mode = 'lines',
        name = 'Tendncia',
        marker_color = '#2ca02c'
    )
)

# Ajustes no layout

Figura.update_layout(
    # Titulo
    title= 'Anlise do fechamento do Bitcoin',
    #Tamanho
    titlefont_size=20,

#Ajustando eixo x
    xaxis = dict(
        title = 'Perodo histrico',
        titlefont_size=14,
        tickfont_size=10,

    ),

    #Ajustando eixo y
    yaxis = dict(
        title = 'Preo fechamento ($)',
        titlefont_size=14,
        tickfont_size=10,


),
#Parametros para legenda
legend =  dict(
    y=1,
    x=1,
    bgcolor = 'rgba(255,0,0,0.2)',
    bordercolor = 'rgba(255,0,0,0.1)',



),


)   
    
```
