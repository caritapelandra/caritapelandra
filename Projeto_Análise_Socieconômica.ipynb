```python
#Importando as Libs

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Lendo a Base de Dados
Base_Dados = pd.read_excel( 'Dados_Pib.xlsx')

#Head
Base_Dados.head(12)
```

```python
#Sistema de grids

#Cor do fundo
Cor_Fundo = '#f5f5f5'

#Criar sistema de grids
Grid_Graficos = sns.FacetGrid(Base_Dados, col='Territorialidades', hue='Territorialidades', col_wrap=4);

#Adicionar um gráfico de linhas em cadas gráfico
Grid_Graficos = Grid_Graficos.map( plt.plot, 'Ano', 'PIB per capita')

#Adicionar uma sombra e ajuste do título
Grid_Graficos = Grid_Graficos.map( plt.fill_between, 'Ano', 'PIB per capita', alpha=0.2). set_titles('{col_name} Territorialidades')

# retirar a palavra territorialidades
Grid_Graficos = Grid_Graficos.set_titles('{col_name}')

#Adicionar subtítulo
Grid_Graficos = Grid_Graficos.fig.suptitle(
    'Evolução da Renda per capita por Estado \n Relatório elaborado no treinamento "Python para Análise de Dados" \n ',
    fontsize=18)

# Ajustando subtítulo
plt.subplots_adjust( top=0.92 )
```
